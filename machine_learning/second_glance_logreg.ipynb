```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, \
                            recall_score, \
                            precision_score, \
                            roc_auc_score, \
                            RocCurveDisplay as roc_curve, \
                            ConfusionMatrixDisplay as conf_mat
```

```python
data_1 = pd.read_csv('AI Generated Essays Dataset.csv')
data_2 = pd.read_csv('train_v2_drcat_02.csv')
X = pd.concat([data_1['text'], data_2['text']])
y = pd.concat([data_1['generated'], data_2['label']])

data = pd.concat([X, y], axis=1)
data.columns = ["text", "label"]
```

```python
estimator = LogisticRegression(solver='lbfgs', C=0.5, max_iter=10000, random_state=0)
h_vect = HashingVectorizer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=0)
```

```python
X_train_pre = h_vect.transform(X_train)
X_test_pre = h_vect.transform(X_test)
```

```python
estimator.fit(X_train_pre, y_train)
```

```python
y_pred = estimator.predict(X_test_pre)

print("=================TEST SCORES==============")
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred)}")
print(f"Recall: {recall_score(y_test, y_pred)}")
print(f"AUC: {roc_auc_score(y_test, y_pred)}")
```
